```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patheffects as pe

# Загружаем данные
import pandas as pd
file_path = "/mnt/data/anxiety_depression_data.csv"
df = pd.read_csv(file_path)

# Определяем показатели и соответствующие столбцы в DataFrame
metrics = {
    "Depression": "Depression_Score",
    "Anxiety": "Anxiety_Score",
    "Stress": "Stress_Level",
    "Self-Esteem": "Self_Esteem_Score",
    "Life Satisfaction": "Life_Satisfaction_Score",
    "Loneliness": "Loneliness_Score"
}

# Цвета для полов
colors = {"Female": "#CCCCCB", "Male": "#595A5A"}

# Средние значения по полу
data = {
    metric: [df[df["Gender"] == "Female"][col].mean(), df[df["Gender"] == "Male"][col].mean()]
    for metric, col in metrics.items()
}

# Создаем позиции для столбиков
x = np.arange(len(metrics))
width = 0.4

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(12, 6), facecolor="#000000")
ax.set_facecolor("#000000")
ax.grid(axis="y", linestyle="--", alpha=0.3, color="gray")

# Добавляем зернистую текстуру фона
ax.patch.set_alpha(0.8)
ax.patch.set_path_effects([pe.SimpleLineShadow(), pe.Normal()])

# Строим столбцы для женщин и мужчин
ax.bar(x - width/2, [data[m][0] for m in metrics], width, label="Women", color=colors["Female"], edgecolor="black")
ax.bar(x + width/2, [data[m][1] for m in metrics], width, label="Men", color=colors["Male"], edgecolor="black")

# Настройки осей и заголовков
ax.set_xticks(x)
ax.set_xticklabels(metrics.keys(), fontsize=10, fontstyle="italic", color="white")
ax.set_title("Psychological Metrics by Gender", fontsize=18, fontweight="bold", fontname="serif", color="white")

# Легенда
ax.legend(title="Gender", title_fontsize=12, fontsize=10, loc="upper right", facecolor="black", edgecolor="white")

# Отображаем график
plt.show()
```
