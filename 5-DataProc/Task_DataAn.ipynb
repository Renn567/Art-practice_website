```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

file_path = file_name
df = pd.read_csv(file_path)

metrics = {
    "Depression": "Depression_Score",
    "Anxiety": "Anxiety_Score",
    "Stress": "Stress_Level",
    "Self-Esteem": "Self_Esteem_Score",
    "Life Satisfaction": "Life_Satisfaction_Score",
    "Loneliness": "Loneliness_Score"
}

colors = {"Female": "#CCCCCB", "Male": "#595A5A"}

data = {
    metric: [df[df["Gender"] == "Female"][col].mean(), df[df["Gender"] == "Male"][col].mean()]
    for metric, col in metrics.items()
}

x = np.arange(len(metrics))
width = 0.4

fig, ax = plt.subplots(figsize=(16, 9), facecolor="#000000")
ax.set_facecolor("#000000")
ax.grid(axis="y", linestyle="--", alpha=0.3, color="gray")

# Строим столбики для женщин и мужчин
ax.bar(x - width/2, [data[m][0] for m in metrics], width, label="Women", color=colors["Female"], edgecolor="black")
ax.bar(x + width/2, [data[m][1] for m in metrics], width, label="Men", color=colors["Male"], edgecolor="black")

# Устанавливаем метки по оси X
ax.set_xticks(x)
ax.set_xticklabels(metrics.keys(), fontsize=10, fontstyle="italic", color="white")

# Устанавливаем заголовок
ax.set_title("Psychological Metrics by Gender", fontsize=18, fontweight="bold", fontstyle="italic", color="white")

# Устанавливаем параметры легенды
ax.legend(title="Gender", title_fontsize=12, fontsize=10, loc="upper right", facecolor="black", edgecolor="white")

# Отображаем график
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = file_name
df = pd.read_csv(file_path)

age_depression = df.groupby("Age")["Depression_Score"].mean()

fig, ax = plt.subplots(figsize=(16, 9))
ax.set_facecolor("#000000")

ax.plot(age_depression.index, age_depression.values, color="#7F7F7F", linewidth=2, marker="o", markerfacecolor="#CCCCCB", markeredgecolor="#7F7F7F")

ax.set_xlabel("Age", fontsize=10, fontstyle="italic", color="white")
ax.set_ylabel("Depression Score", fontsize=10, fontstyle="italic", color="white")
ax.set_title("Average Depression by Age", fontsize=18, fontweight="bold", color="white", fontstyle="italic")

ax.spines["bottom"].set_color("white")
ax.spines["left"].set_color("white")
ax.spines["top"].set_color("white")
ax.spines["right"].set_color("white")

ax.tick_params(axis="both", colors="white")
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_color("white")
    label.set_fontstyle('italic')

fig.patch.set_facecolor('black')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = file_name
df = pd.read_csv(file_path)

df_female = df[df['Gender'] == 'Female']
df_male = df[df['Gender'] == 'Male']

age_depression_female = df_female.groupby("Age")["Depression_Score"].mean()
age_depression_male = df_male.groupby("Age")["Depression_Score"].mean()

fig, ax = plt.subplots(figsize=(16, 9))
ax.set_facecolor("#000000")

ax.plot(age_depression_female.index, age_depression_female.values, color="#CCCCCB", linewidth=2, marker="o", markerfacecolor="#CCCCCB", markeredgecolor="#CCCCCB", label="Female")
ax.plot(age_depression_male.index, age_depression_male.values, color="#676868", linewidth=2, marker="o", markerfacecolor="#676868", markeredgecolor="#676868", label="Male")

ax.set_xlabel("Age", fontsize=10, fontstyle="italic", color="white")
ax.set_ylabel("Depression Score", fontsize=10, fontstyle="italic", color="white")
ax.set_title("Average Depression by Age (Female vs Male)", fontsize=18, fontweight="bold", color="white")

ax.spines["bottom"].set_color("white")
ax.spines["left"].set_color("white")
ax.spines["top"].set_color("white")
ax.spines["right"].set_color("white")

ax.tick_params(axis="both", colors="white")
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_color("white")

fig.patch.set_facecolor('black')

ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = file_name
df = pd.read_csv(file_path)

employment_counts = df['Employment_Status'].value_counts()

labels = employment_counts.index
sizes = employment_counts.values

fig, ax = plt.subplots(figsize=(16, 9), facecolor='black')

ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'color': 'white'},
       colors=['#7F7F7F', '#CCCCCB', '#676868', '#999999', '#444444'])

ax.set_title('Employment Status Distribution', fontsize=18, fontweight='bold', color='white', fontstyle='italic')

fig.patch.set_facecolor('black')
ax.set_facecolor('black')

ax.set_xlabel('')
ax.set_ylabel('')

for label in ax.texts:
    label.set_fontstyle('italic')
    label.set_fontsize(10)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = file_name
df = pd.read_csv(file_path)

correlation_factors = {
    'Loneliness_Score': df['Loneliness_Score'].corr(df['Depression_Score']),
    'Life_Satisfaction_Score': df['Life_Satisfaction_Score'].corr(df['Depression_Score']),
    'Self_Esteem_Score': df['Self_Esteem_Score'].corr(df['Depression_Score']),
    'Work_Stress': df['Work_Stress'].corr(df['Depression_Score']),
    'Financial_Stress': df['Financial_Stress'].corr(df['Depression_Score'])
}

correlation_df = pd.DataFrame(list(correlation_factors.items()), columns=["Factor", "Correlation"])
correlation_df["Abs_Correlation"] = correlation_df["Correlation"].abs()

labels = correlation_df["Factor"]
sizes = correlation_df["Abs_Correlation"]

fig, ax = plt.subplots(figsize=(16, 9), facecolor='black')

ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, textprops={'color': 'white'},
       colors=['#7F7F7F', '#CCCCCB', '#676868', '#999999', '#444444'])

ax.set_title('Factors Affecting Depression', fontsize=18, fontweight='bold', color='white', fontstyle='italic')

fig.patch.set_facecolor('black')
ax.set_facecolor('black')

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_color("white")

for label in ax.texts:
    label.set_fontstyle('italic')

plt.tight_layout()
plt.show()
```
